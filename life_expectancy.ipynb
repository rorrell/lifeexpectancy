# How Does Life Expectancy Change Over Time and How Does It Compare Between Men and Women?

*Setup*

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Sources/NCHS_-_Death_rates_and_life_expectancy_at_birth.csv")
print(df.shape)
df.head()
```

*Customize grids*

```python
sns.set_style('whitegrid')
```

*Visualization*

```python
# Create filtered dataframes
men = df[(df["Sex"] == "Male") & (df["Race"] == "All Races")]
women = df[(df["Sex"] == "Female") & (df["Race"] == "All Races")]

# Create line plots
plt.plot(men["Year"], men["Average Life Expectancy (Years)"], color="Blue", label="Men")
plt.plot(women["Year"], women["Average Life Expectancy (Years)"], color="Red", label="Women")

# Customize chart
plt.legend()
plt.title("Life Expectancy in Years for Men and Women From 1900-2018")
plt.xlabel("Year")
plt.ylabel("Life Expectancy (Years)")
plt.show()
```

You can conclude from the above that there was a big dip around the Spanish flu and that the gap between men and women started widening not long after 1940 and has stayed mostly stable since around 1960.  Generally there has been an upward trend, though that may be starting to plateau.  Women have almost always lived longer than men during this period of time.
